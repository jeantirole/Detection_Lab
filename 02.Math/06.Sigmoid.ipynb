```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x, diff):
    if diff:
        # f(x) x (1-f(x))
        return (1 / (1 + np.exp(-x))) * (1 - (1 / (1 + np.exp(-x))))

    else:
        return 1 / (1 + np.exp(-x))

# Example usage:
x = np.array([-1, 0, 1])
print(sigmoid(x,False))
print(sigmoid(x,True))
```

```python
# Generate x values
x_values = np.linspace(-10, 10, 100)  # Generate 100 values from -10 to 10

# Calculate corresponding y values using the sigmoid function
y_values = sigmoid(x_values,False)

# Plot the sigmoid function
plt.plot(x_values, y_values)
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.grid(True)
plt.show()
```

```python
# Generate x values
x_values = np.linspace(-10, 10, 100)  # Generate 100 values from -10 to 10

# Calculate corresponding y values using the sigmoid function
y_values = sigmoid(x_values,True)

# Plot the sigmoid function
plt.plot(x_values, y_values)
plt.title('Sigmoid diff Function')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.grid(True)
plt.show()
```
