# Dot Product

- topic : 두 벡터가 주어졌을 때, theta 구하기

- dot_product(vector a, vector b ) = | a | x | b | x cos theta
-                                  = a1 x a2 + b1 x b2 <- 코사인법칙으로 증명가능
- cos theta =  dot_product(vector a, vector b)  / (|a| x |b|) 

```python
import math
import numpy as np

vector_a = [4,2]
vector_b = [5,0]
```

```python
# dot product ! 

def dot_product_py(vector_a, vector_b):
    dot_product_result = 0
    for i in range(2):
        a_ = vector_a[i]
        b_ = vector_b[i]
        
        dot_r = a_ * b_
        dot_product_result += dot_r

    return dot_product_result    
```

```python
# vector norm = np.sqrt(sum([i*i for i in vector_a]))

cos_theta = dot_product_py(vector_a,vector_b) / (np.linalg.norm(vector_a) * np.linalg.norm(vector_b))

# Highlight the angle theta
theta_rad = np.arccos(np.clip(cos_theta, -1.0, 1.0))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

def dot_product_py(vector_a, vector_b):
    dot_product_result = 0
    for i in range(2):
        a_ = vector_a[i]
        b_ = vector_b[i]
        
        dot_r = a_ * b_
        dot_product_result += dot_r

    return dot_product_result

# Vectors
vector_a = np.array([4, 2])
vector_b = np.array([5, 0])

# Calculate cosine of the angle
def update(frame):
    plt.clf()  # Clear the previous frame
    theta = np.radians(frame)
    
    # Rotate vector_b by angle theta
    rotated_vector_b = np.array([
        vector_b[0] * np.cos(theta) - vector_b[1] * np.sin(theta),
        vector_b[0] * np.sin(theta) + vector_b[1] * np.cos(theta)
    ])

    # Plot the vectors
    plt.quiver(0, 0, vector_a[0], vector_a[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vector A')
    plt.quiver(0, 0, rotated_vector_b[0], rotated_vector_b[1], angles='xy', scale_units='xy', scale=1, color='b', label='Rotated Vector B')

    # Highlight the angle theta
    plt.annotate("", xy=(rotated_vector_b[0], rotated_vector_b[1]), xytext=(vector_a[0], vector_a[1]), arrowprops=dict(arrowstyle="->", color='g'))
    plt.annotate(f'{frame} degrees', xy=(0.5, 0.2), fontsize=12, color='g')

    # Set the aspect ratio of the plot to be equal
    plt.gca().set_aspect('equal', adjustable='box')

    # Set axis limits
    plt.xlim(-6, 6)
    plt.ylim(-6, 6)

    # Add labels and legend
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.title('Animation of Rotating Vectors')
    plt.legend()

# Create an animation
fig, ax = plt.subplots()
animation = FuncAnimation(fig, update, frames=np.arange(0, 360, 5), interval=100)

# Show the animation
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate dot product
def dot_product(a,b,theta):
    size_a  = np.sqrt(sum([i*i for i in a]))
    size_b  = np.sqrt(sum([i*i for i in b]))
    cosine_theta = math.cos(theta)
    result_vector_dot_product = size_a * size_b * cosine_theta
    return result_vector_dot_product

# Vectors
vector_a = np.array([4, 2])
vector_b = np.array([5, 0])
degree = 30
theta = math.pi * (degree / 180)

# Calculate dot product
result = dot_product(vector_a, vector_b, theta)

# Calculate projection
projected_vector = np.sqrt(sum([i*i for i in vector_a])) * math.cos(theta)  

# Plot the vectors
plt.figure(figsize=(8, 8))
plt.quiver(0, 0, vector_a[0], vector_a[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vector A')
plt.quiver(0, 0, vector_b[0], vector_b[1], angles='xy', scale_units='xy', scale=1, color='b', label='Vector B')

# Highlight the angle theta
theta_rad = np.arccos(result / (np.linalg.norm(vector_a) * np.linalg.norm(vector_b)))
#plt.annotate("", xy=(projected_vector[0], projected_vector[1]), xytext=(vector_a[0], vector_a[1]), arrowprops=dict(arrowstyle="->", color='g'))
plt.annotate(f'{degree} degrees', xy=(0.5, 0.2), fontsize=12, color='g')

# Set the aspect ratio of the plot to be equal
plt.gca().set_aspect('equal', adjustable='box')

# Set axis limits
plt.xlim(0, 8)
plt.ylim(0, 3)

# Add labels and legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Visualization of Vectors and Dot Product')
plt.legend()

# Show the plot
plt.grid()
plt.show()
```
