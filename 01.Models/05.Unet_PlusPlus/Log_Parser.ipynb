```python
import numpy as np
import matplotlib.pyplot as plt
import os 
```

```python
log_ = sorted(os.listdir("./01.log"))[-1]
f = open(os.path.join("./01.log", log_))
lines = f.readlines()
```

```python
print(log_)
```

```python
for line in lines:
    print(line.split())
```

```python
logs_ = {
    'epochs' : [],
    'iterations' : [],
    'losses' : [],
    'dices' : []
}

for line in lines:
    
    dice = line.split()[-1]
    loss = line.split()[-3]
    iter = line.split()[-5]
    epoch = line.split()[-8]
    
    logs_["dices"].append(dice)
    logs_["losses"].append(loss)
    logs_["iterations"].append(iter)
    logs_["epochs"].append(epoch)        
```

```python
losses = [float(i) for i in logs_['losses']]
dices =  [float(i) for i in logs_['dices']]
```

```python
# Plot loss over time
plt.figure(figsize=(20, 12))
plt.plot(losses, marker='o', linestyle='-')
plt.title('Loss over Time')
plt.xlabel('Iterations')
plt.ylabel('Loss')
```

```python
import matplotlib.pyplot as plt

# Assuming 'losses' is a list containing loss values

# Plot loss over time as a scatter plot
plt.figure(figsize=(20, 12))
plt.scatter(range(len(losses)), losses, marker='o', color='blue')
plt.title('Loss over Time')
plt.xlabel('Iterations')
plt.ylabel('Loss')
plt.grid(True)  # Add grid for better visualization
plt.show()
```

```python
# Plot loss over time
plt.figure(figsize=(20, 12))
plt.plot(dices, marker='o', linestyle='-')
plt.title('Loss over Time')
plt.xlabel('Iterations')
plt.ylabel('dice')
```
