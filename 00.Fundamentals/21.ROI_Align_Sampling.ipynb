```python
# Bi-Linear interpolation 

'''
formula : 

p = (d1 / (d1+d2)) * p1 + (d2 / (d1+d2)) * p2  


'''
```

```python
def interpolated_point(p,p1,p2):
    d1 = abs(p1-p)
    d2 = abs(p2-p)
    p = (d1 / (d1+d2)) * p1 + (d2 / (d1+d2)) * p2  
    
    return p 
```
