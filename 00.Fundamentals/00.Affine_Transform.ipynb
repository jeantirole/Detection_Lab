```python
import cv2
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
import math 
```

```python
#-------------------------------------
# params
theta = 30
theta = theta * (np.pi / 180)
print(theta)

tx = 0
ty = 0
#-------------------------------------

matrix = np.array([
    [np.cos(theta) , -np.sin(theta), tx],
    [np.sin(theta),  np.cos(theta), ty],
    [0 , 0, 1 ]])

matrix = np.transpose(matrix)

print(matrix)
print(matrix.shape)
```

```python
point= np.array([[12.5, 5,1]])
point.shape
```

```python
point @ matrix
```

```python
fig, ax = plt.subplots(figsize=(6, 6))

rotated_point = point @ matrix

ax.add_patch(
    patches.FancyArrow(shape="full", x=0,y=0,dx=point[0][0],dy=point[0][1],head_length=1,head_width=1.5)
)
ax.add_patch(
    patches.FancyArrow(shape="full", x=0,y=0,dx=rotated_point[0][0],dy=rotated_point[0][1],head_length=1,head_width=1.5,color='r')
)
ax.set_xlim(0,20)
ax.set_ylim(0,20)
```
